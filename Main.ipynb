```python
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
    "https://www.wipro.com/about-us/",
    "https://www.tcs.com/",  # Add other URLs here
    "https://www.infosys.com/",
    "https://www.icicibank.com/",
    "https://www.capgemini.com/",
    "https://www.nttdata.com/",
    "https://www.hcltech.com/",
    "https://www.techmahindra.com/",
    "https://www.lntinfotech.com/",
    "https://www.mindtree.com/",
    "https://www.mphasis.com/",
    "https://www.persistent.com/",
    "https://www.cognizant.com/",
    "https://www.accenture.com/",
    "https://www.ibm.com/in-en/",
    "https://www.oracle.com/in/",
    "https://about.google/intl/en-in",
    "https://aws.amazon.com/?nc2=h_lg",
    "https://www.cisco.com/c/en_in",
    "https://www.qualcomm.com/",
    "https://www.vmware.com/in",
    "https://www.redhat.com/en",
    "https://www.capgemini.com/in-en",
    "https://atos.net/en/india",
    "https://www.cgi.com/india",
    "https://www.hexaware.com",
    "https://www.zensar.com",
    "https://www.sonata-software.com",
    "https://www.niit.com/india/",
    "https://www.cerner.com",
    "https://www.ust.com",
    "https://www.3i-infotech.com",
    "https://www.hexaware.com"
]

# Set the User-Agent header to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.134 Safari/537.36"
}

# Function to scrape the company name from a single URL
def scrape_company_name(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url, headers=headers)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(response.text, "html.parser")
            
            # Find the company name in the title tag
            title_tag = soup.find("title")
            if title_tag:
                title_text = title_tag.get_text()
                # Extract the company name from the title text
                company_name = title_text.split(" - ")[0].strip()
                return company_name
            else:
                print(f"Title tag not found for {url}.")
                return None
        else:
            print(f"Failed to retrieve {url}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None

# List to store the company names
company_names = []

# Loop over the list of URLs and scrape the company name from each
for url in urls:
    company_name = scrape_company_name(url)
    if company_name:  # Only add if company name was successfully retrieved
        company_names.append(company_name)

# Output the list of company names
print("List of Company Names:")
print(company_names)
```

```python
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
    "https://www.wipro.com/about-us/",
    "https://www.tcs.com/",  # Add other URLs here
    "https://www.infosys.com/",
    "https://www.icicibank.com/",
    "https://www.capgemini.com/",
    "https://www.nttdata.com/",
    "https://www.hcltech.com/",
    "https://www.techmahindra.com/",
    "https://www.lntinfotech.com/",
    "https://www.mindtree.com/",
    "https://www.mphasis.com/",
    "https://www.persistent.com/",
    "https://www.cognizant.com/",
    "https://www.accenture.com/",
    "https://www.ibm.com/in-en/",
    "https://www.oracle.com/in/",
    "https://about.google/intl/en-in",
    "https://aws.amazon.com/?nc2=h_lg",
    "https://www.cisco.com/c/en_in",
    "https://www.qualcomm.com/",
    "https://www.vmware.com/in",
    "https://www.redhat.com/en",
    "https://www.capgemini.com/in-en",
    "https://atos.net/en/india",
    "https://www.cgi.com/india",
    "https://www.hexaware.com",
    "https://www.zensar.com",
    "https://www.sonata-software.com",
    "https://www.niit.com/india/",
    "https://www.cerner.com",
    "https://www.ust.com",
    "https://www.3i-infotech.com",
    "https://www.hexaware.com"
]

# Set the User-Agent header to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.134 Safari/537.36"
}

# Function to scrape the website URL from a single URL
def scrape_website_url(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url, headers=headers)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(response.text, "html.parser")
            
            # Find the <link> tag with rel="canonical" to get the main website URL
            canonical_tag = soup.find("link", rel="canonical")
            if canonical_tag and canonical_tag.get("href"):
                website_url = canonical_tag.get("href")
                return website_url
            else:
                print(f"Canonical link not found for {url}.")
                return url  # Return the original URL if no canonical link is found
        else:
            print(f"Failed to retrieve {url}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None

# List to store the website URLs
website_urls = []

# Loop over the list of URLs and scrape the website URL from each
for url in urls:
    website_url = scrape_website_url(url)
    if website_url:  # Only add if website URL was successfully retrieved
        website_urls.append(website_url)

# Output the list of website URLs
print("List of Website URLs:")
print(website_urls)
```
